```python
import os
```

```python
def find_my_files(pattern,directory):
    aim_files_list=[]
    for  dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            if pattern in filename:
                aim_files_list.append(os.path.join(dirpath,filename))
    return aim_files_list
```
